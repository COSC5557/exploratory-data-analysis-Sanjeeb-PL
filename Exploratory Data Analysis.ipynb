Lets normailze the data and see the histogram

```python
from sklearn import preprocessing
# normalize each column individually
dfn = preprocessing.normalize(df, axis= 0)
# convert the normalized array back to dataframe
dfn = pd.DataFrame(dfn, columns=df.columns)
dfn.hist(bins=50, figsize= (20,15))
#plt.plot(dfn, density(dfn))
plt.tight_layout()
plt.show()
```
