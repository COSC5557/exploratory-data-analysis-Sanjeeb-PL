```python
df2 = df.groupby('quality')[['residual sugar','fixed acidity', 'citric acid',
       'chlorides', 'density','pH', 'alcohol']].mean().sort_values(by='density', ascending= False)
df3 = df2.transpose()
df3.plot()
```

```python
#df4 = df.groupby('quality')[['residual sugar', 'alcohol']].mean().sort_values(by='alcohol', ascending= False)
df4 = df.groupby('fixed acidity')[['density', 'fixed acidity']].mean()
print(df4)
#df4.plot()
plt.plot(df4['fixed acidity'],df4['density'], marker = '*', linestyle= '-')
plt.xlabel('fixed acidity')
plt.ylabel('Desnsity')
plt.title('Plot of fixed acidity vs Density')
plt.grid(True)
plt.show()
```
